Now lets define a Threshold to help classify cars as either High-performance/ luxury cars or not. for simplicity we will call the column 'luxury'

```python
lamborghini_average_price       = df.loc[df['brand'] == 'aston-martin', 'price_in_euro'].mean()
audi_average_price              = df.loc[df['brand'] == 'audi',         'price_in_euro'].mean()
bentley_average_price           = df.loc[df['brand'] == 'bentley',      'price_in_euro'].mean()
ferrari_average_price           = df.loc[df['brand'] == 'ferrari',      'price_in_euro'].mean()
lamborghini_average_price       = df.loc[df['brand'] == 'lamborghini',  'price_in_euro'].mean()
maserati_average_price          = df.loc[df['brand'] == 'maserati',     'price_in_euro'].mean()
porsche_average_price           = df.loc[df['brand'] == 'porsche',      'price_in_euro'].mean()
```

```python
# This is a custom made function for the fuel consumption column to extract only the numbers
# and save them in the cell
# For Example:
# '10,9 l/100' km is converted into '10.9' as a numeric value
def convert_to_numeric(value):
    try:
        if value.strip():  # Check if the string is not empty
            # Extract numeric part, split by comma, replace ',' with '.' and concatenate
            return float('.'.join(value.split()[0].split(',')[0:2]))
        else:
            return np.nan  # Set empty cells to NaN
    except AttributeError:
        return value  # Return the original value if it's already numeric
    except Exception as e:
        # print(f"Error: {e}")
        return None

df['fuel_consumption_l_100km'] = df['fuel_consumption_l_100km'].apply(convert_to_numeric)
```

```python
# Violin plots for fuel consumption and mileage by fuel type

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.violinplot(x='fuel_type', y='fuel_consumption_l_100km', data=df)
plt.title('Violin Plot: Fuel Consumption by Fuel Type')

plt.subplot(1, 2, 2)
sns.violinplot(x='fuel_type', y='mileage_in_km', data=df)
plt.title('Violin Plot: Mileage by Fuel Type')
plt.ylim(0, 1000000)  # Set y-axis limits from 0 to 1000000 to zoom in a bit

plt.tight_layout()
plt.show()

# Here these 2 are similar to the BoxPlots, but we can actually see the distribution and weights for different fuel consumption and mileage values.

# Fuel Consumption Plot:
# We can see that Diesel has a lower fuel consumption. Petrol also has a more weighter fuel consumption at the top of the plot,
# meaning There are more Petrol cars with high fuel coonsumption than Diesel cars.
# The same goes for the bottom of the plot. We can see that there are more Diesel Cars with low fuel consumption than Petrol Cars.

# Mileage Plot
# We can clearly see that there are way more Petrol cars than Diesel cars, that have a low mileage,
# meaning that there is a little amount of Diesel cars with low mileage.
# We can also see that Diesel cars can reach higher mileages than Petrol cars.
```

```python
# Histograms for fuel consumption and mileage by fuel type
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df[df['fuel_type'] == 'Diesel']['fuel_consumption_l_100km'], color='blue', label='Diesel', kde=True)
sns.histplot(df[df['fuel_type'] == 'Petrol']['fuel_consumption_l_100km'], color='red', label='Petrol', kde=True)
plt.title('Histogram: Fuel Consumption Distribution by Fuel Type')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(df[df['fuel_type'] == 'Diesel']['mileage_in_km'], color='blue', label='Diesel', kde=True)
sns.histplot(df[df['fuel_type'] == 'Petrol']['mileage_in_km'], color='red', label='Petrol', kde=True)
plt.title('Histogram: Mileage Distribution by Fuel Type')
plt.legend()
plt.ylim(0, 6000)  # Set y-axis limits from 0 to 6000 to zoom in using the y-axis
plt.xlim(0, 500000)  # Set x-axis limits from 0 to 100000

plt.tight_layout()
plt.show()

# Fuel Consumption Histogram:
# This chart clearly shows that Diesel-Fueled-Cars have a much lower average fuel consumption than
# Petrol-Fueled-Cars.
# We can see this through the blue line.

# Mileage Histogram:
# We see that the red line (Petrol-Fueled-Cars) have lot more cars with a low mileage than the blue line (Diesel-Fueled-Cars)
```
