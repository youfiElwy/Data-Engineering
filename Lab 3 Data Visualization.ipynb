```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
def get_path(dataset_name,env_name='colab'):
    """
    This function is used to return the path of the dataset you want to use.

    @params:
    dataset_name: the name of the dataset.
    env_name: it has two values either local, or colab the default is colab
    """
    prefix = 'https://raw.githubusercontent.com/mohamed-ashry7/Data-Engineering-Lab/main/Datasets/'
    if env_name == 'colab':
        return prefix+dataset_name
    else:
        return f'../Datasets/{dataset_name}'
```

### **Bar Graph**

```python
movie_scores = pd.read_csv(get_path('movie_scores.csv'))
movie_scores
```

barplots are for categorical x axis values.

histogram are for continues x axis values.

scatter plot to find patterns between 2 continues values.

bubble plot is same as scatter but the dots have sizes

box plot is for mean/median and outliers


```python
movies_fig = plt.figure(figsize=(10, 5), dpi=300)
ax = movies_fig.add_subplot(1,1,1)
x = np.arange(1,len(movie_scores.Tomatometer)+1)
x
width = 0.3
ax.bar(height = movie_scores.Tomatometer , width=width,  x = x-0.3/2)
ax.bar(height = movie_scores.AudienceScore,width=width, x = x+0.3/2)
ax.set_xticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_xticks(x)
```

### For horizontal bar just flip (x,y), (Width, Height) and so on.

```python
movies_fig = plt.figure(figsize=(10, 5), dpi=300)
ax = movies_fig.add_subplot(1,1,1)
y = np.arange(1,len(movie_scores.Tomatometer)+1)
width = 0.2
ax.barh(width= movie_scores.Tomatometer , height=width,  y = y-0.3/2)
ax.barh(width =movie_scores.AudienceScore,height=width, y = y+0.3/2)
ax.set_yticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_yticks(y)
```

### **Histogram**

The difference between Histogram and barplot is that barplot comparing many columns/attributes together.  
On the other side, Histogram shows the distribution of values for a specific column.

```python
hist_fig, ax = plt.subplots(1,1)
ax.hist(np.random.randn(100),bins=50)
ax.hist(np.random.randn(100),bins=50)
```

```python
df_titanic = pd.read_csv(get_path('titanic.csv'),index_col=0)
df_titanic.head()
```

```python
hist_fig, ax = plt.subplots(1,1)
max_age = max(df_titanic.Age.values)
min_age = min(df_titanic.Age.values)
age_bins= int((max_age - min_age)//10)
ax.hist(df_titanic.Age.values, bins=age_bins, color='k')
ax.grid()
```

### ***Bubble Plot***

Bubble plot is just a scatter plot with the ability to add another 2 dimensions which can be viewed as the scale s and the color c

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
sp = ax.scatter(np.random.randn(20),np.random.randn(20),s=np.abs(np.random.randn(20))*100,alpha=0.5,c= np.abs([np.random.randn(20)]))
fig.colorbar(sp)
```

### **Distribution Plots**

Distribution plots, display the distribution for one variable.

The below dataset describes the driver's status, when the car crashed in a certain state in the United States. Our role is trying to identify several factors:

1. How likely it is that the driver was not distracted when the car crashed?
2. What is the correlation between the driver not being distracted and speeding when the car crashed?
3. What is the relation between the car crashing and every two factors in the dataset?

```python
df_car_crashes = sns.load_dataset("car_crashes")#seaborn has built in functions to retrieve datasets
df_car_crashes.head()
```

```python
sns.distplot(df_car_crashes['not_distracted'],bins=20,kde=True)
#plots the kernel density estimate
```

### **Join Plots**

Join plots show the correlation between two variables. In the below chart, they show the correlation of the accident taking place such that the person was both speeding and not distracted.

```python
sns.jointplot(df_car_crashes['not_distracted'],df_car_crashes['speeding'],kind='reg')
```
